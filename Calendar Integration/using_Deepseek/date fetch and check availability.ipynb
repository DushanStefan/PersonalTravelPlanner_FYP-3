```python
import re
import dateparser
from datetime import datetime, timedelta
from googleapiclient.discovery import build
from google.oauth2 import service_account
from langchain_community.llms import Ollama
from langchain import PromptTemplate, LLMChain

# Google Calendar API Authentication
SCOPES = ['https://www.googleapis.com/auth/calendar.readonly']
SERVICE_ACCOUNT_FILE = "/content/calendarcolab-ServiceAccount.json"  # Ensure this file is uploaded

# Authenticate and build Google Calendar service
credentials = service_account.Credentials.from_service_account_file(SERVICE_ACCOUNT_FILE, scopes=SCOPES)
calendar_service = build("calendar", "v3", credentials=credentials)

# Get Today's Date
today_date = datetime.today().strftime("%B %d, %Y")  # Example: "February 19, 2025"

# Load DeepSeek R1 in Ollama
llm = Ollama(model="deepseek-r1:7b")

# Define Prompt Template
template = f"""
Extract travel-related information from the query.
If the start date is not mentioned, use today's date: {today_date}.

Strictly return a valid JSON object with the following keys:
- 'start_date'
- 'end_date'
- 'duration'

If a relative date (e.g., "next Monday", "in two weeks", "end of the month") is mentioned, convert it into a standard date format.

Do not include any explanation, reasoning, or extra text.

Query: {{query}}

Output:
"""

prompt = PromptTemplate(template=template, input_variables=["query"])
chain = LLMChain(llm=llm, prompt=prompt)

# Function to clean LangChain response
def clean_response(response):
    return re.sub(r"<think>.*?</think>", "", response, flags=re.DOTALL).strip()

# Function to parse JSON response
def parse_dates(response_json):
    try:
        date_pattern = r'("start_date":\s*"([^"]+)")|("end_date":\s*"([^"]+)")|("duration":\s*"([^"]+)")'
        matches = re.findall(date_pattern, response_json)

        structured_data = {}
        for match in matches:
            if match[0]: structured_data["start_date"] = match[1]
            if match[2]: structured_data["end_date"] = match[3]
            if match[4]: structured_data["duration"] = match[5]

        return structured_data
    except Exception as e:
        return {"error": f"Error parsing the JSON response: {e}"}

# Convert human-readable dates to actual dates
def format_date(date_str):
    parsed_date = dateparser.parse(date_str)
    if parsed_date:
        return parsed_date.strftime("%Y-%m-%d")  # Convert to ISO format
    return date_str

# Function to check availability in Google Calendar
def check_availability(start_date, end_date):
    start_iso = f"{start_date}T00:00:00Z"  # Start of day
    end_iso = f"{end_date}T23:59:59Z"  # End of day

    events_result = calendar_service.events().list(
        calendarId='primary',
        timeMin=start_iso,
        timeMax=end_iso,
        singleEvents=True,
        orderBy='startTime'
    ).execute()

    events = events_result.get("items", [])

    available_slots = []
    unavailable_slots = []

    # If no events, user is fully available
    if not events:
        return {"availability": "Fully Available", "slots": []}

    # Check for unavailable slots
    for event in events:
        start = event["start"].get("dateTime", event["start"].get("date"))
        end = event["end"].get("dateTime", event["end"].get("date"))
        unavailable_slots.append({"start": start, "end": end})

    # Now check for available slots between events
    # Sort events based on start time
    events.sort(key=lambda e: e['start'].get('dateTime', e['start'].get('date')))

    # Find available slots
    for i in range(1, len(events)):
        prev_end = events[i-1]["end"].get("dateTime", events[i-1]["end"].get("date"))
        curr_start = events[i]["start"].get("dateTime", events[i]["start"].get("date"))

        # Compare previous event end time with current event start time
        prev_end = dateparser.parse(prev_end)
        curr_start = dateparser.parse(curr_start)

        # If there's a gap between events, mark it as an available slot
        if curr_start > prev_end:
            available_slots.append({"start": prev_end.strftime("%Y-%m-%d %H:%M:%S"), "end": curr_start.strftime("%Y-%m-%d %H:%M:%S")})

    return {
        "availability": "Partially Available" if unavailable_slots else "Fully Available",
        "unavailable_slots": unavailable_slots,
        "available_slots": available_slots
    }

# Function to process user input query
def process_user_query(query):
    # Get response from LangChain
    response = chain.run(query)
    cleaned_response = clean_response(response)
    structured_response = parse_dates(cleaned_response)

    # Format Dates
    start_date = format_date(structured_response.get("start_date", today_date))
    end_date = format_date(structured_response.get("end_date", today_date))

    # Check Calendar Availability
    availability_result = check_availability(start_date, end_date)

    return {
        "query": query,
        "travel_dates": {"start_date": start_date, "end_date": end_date},
        "availability": availability_result['availability'],
        "unavailable_slots": availability_result.get('unavailable_slots', []),
        "available_slots": availability_result.get('available_slots', [])
    }

# Example: Input query from user
query = input("Enter your travel query: ")

# Process the query and display the results
result = process_user_query(query)

print(f"🔹 **Query:** {result['query']}")
print(f"📅 **Travel Dates:** {result['travel_dates']['start_date']} to {result['travel_dates']['end_date']}")
print(f"✅ **Availability:** {result['availability']}")

if result['unavailable_slots']:
    print(f"⛔ **Unavailable Slots:**")
    for slot in result['unavailable_slots']:
        print(f"  - {slot['start']} to {slot['end']}")
else:
    print("⛔ **Unavailable Slots:** None")

if result['available_slots']:
    print(f"✅ **Available Slots:**")
    for slot in result['available_slots']:
        print(f"  - {slot['start']} to {slot['end']}")
else:
    print("✅ **Available Slots:** None")
```

```python
import re
import dateparser
from datetime import datetime, timedelta
from googleapiclient.discovery import build
from google.oauth2 import service_account
from langchain_community.llms import Ollama
from langchain import PromptTemplate, LLMChain

# Google Calendar API Authentication
SCOPES = ['https://www.googleapis.com/auth/calendar.readonly']
SERVICE_ACCOUNT_FILE = "/content/calendarcolab-ServiceAccount.json"  # Ensure this file is uploaded

# Authenticate and build Google Calendar service
credentials = service_account.Credentials.from_service_account_file(SERVICE_ACCOUNT_FILE, scopes=SCOPES)
calendar_service = build("calendar", "v3", credentials=credentials)

# Get Today's Date
today_date = datetime.today().strftime("%B %d, %Y")  # Example: "February 19, 2025"

# Load DeepSeek R1 in Ollama
llm = Ollama(model="deepseek-r1:7b")

# Define Prompt Template
template = f"""
Extract travel-related information from the query.
If the start date is not mentioned, use today's date: {today_date}.

Strictly return a valid JSON object with the following keys:
- 'start_date'
- 'end_date'
- 'duration'

If a relative date (e.g., "next Monday", "in two weeks", "end of the month") is mentioned, convert it into a standard date format.

Do not include any explanation, reasoning, or extra text.

Query: {{query}}

Output:
"""

prompt = PromptTemplate(template=template, input_variables=["query"])
chain = LLMChain(llm=llm, prompt=prompt)

# Function to clean LangChain response
def clean_response(response):
    return re.sub(r"<think>.*?</think>", "", response, flags=re.DOTALL).strip()

# Function to parse JSON response
def parse_dates(response_json):
    try:
        date_pattern = r'("start_date":\s*"([^"]+)")|("end_date":\s*"([^"]+)")|("duration":\s*"([^"]+)")'
        matches = re.findall(date_pattern, response_json)

        structured_data = {}
        for match in matches:
            if match[0]: structured_data["start_date"] = match[1]
            if match[2]: structured_data["end_date"] = match[3]
            if match[4]: structured_data["duration"] = match[5]

        return structured_data
    except Exception as e:
        return {"error": f"Error parsing the JSON response: {e}"}

# Convert human-readable dates to actual dates
def format_date(date_str):
    parsed_date = dateparser.parse(date_str)
    if parsed_date:
        return parsed_date.strftime("%Y-%m-%d")  # Convert to ISO format
    return date_str

# Function to check availability in Google Calendar
def check_availability(start_date, end_date):
    start_iso = f"{start_date}T00:00:00Z"  # Start of day (midnight)
    end_iso = f"{end_date}T23:59:59Z"  # End of day (11:59 PM)

    events_result = calendar_service.events().list(
        calendarId='primary',
        timeMin=start_iso,
        timeMax=end_iso,
        singleEvents=True,
        orderBy='startTime'
    ).execute()

    events = events_result.get("items", [])
    unavailable_slots = []
    partially_available = False

    # If no events, user is fully available
    if not events:
        return {"availability": "Fully Available", "unavailable_slots": []}

    for event in events:
        start = event["start"].get("dateTime", event["start"].get("date"))
        end = event["end"].get("dateTime", event["end"].get("date"))

        # Convert to datetime objects for comparison
        start_dt = dateparser.parse(start)
        end_dt = dateparser.parse(end)
        requested_start = dateparser.parse(f"{start_date}T00:00:00")
        requested_end = dateparser.parse(f"{end_date}T23:59:59")

        # Check if the event fully or partially overlaps the requested period
        if start_dt < requested_end and end_dt > requested_start:
            unavailable_slots.append({"start": start, "end": end})

            # If the event is fully within the requested period
            if start_dt >= requested_start and end_dt <= requested_end:
                continue  # This is a fully blocked slot, so continue checking for other slots

            # If the event partially overlaps, mark partially available
            partially_available = True

    # Determine overall availability
    if unavailable_slots:
        if partially_available:
            availability_status = "Partially Available"
        else:
            availability_status = "Unavailable"
    else:
        availability_status = "Fully Available"

    return {
        "availability": availability_status,
        "unavailable_slots": unavailable_slots
    }

# Function to process user input query
def process_user_query(query):
    # Get response from LangChain
    response = chain.run(query)
    cleaned_response = clean_response(response)
    structured_response = parse_dates(cleaned_response)

    # Format Dates
    start_date = format_date(structured_response.get("start_date", today_date))
    end_date = format_date(structured_response.get("end_date", today_date))

    # Check Calendar Availability
    availability_result = check_availability(start_date, end_date)

    return {
        "query": query,
        "travel_dates": {"start_date": start_date, "end_date": end_date},
        "availability": availability_result['availability'],
        "unavailable_slots": availability_result.get('unavailable_slots', [])
    }

# Example: Input query from user
query = input("Enter your travel query: ")

# Process the query and display the results
result = process_user_query(query)

print(f"🔹 **Query:** {result['query']}")
print(f"📅 **Travel Dates:** {result['travel_dates']['start_date']} to {result['travel_dates']['end_date']}")
print(f"✅ **Availability:** {result['availability']}")

if result['unavailable_slots']:
    print(f"⛔ **Unavailable Slots:**")
    for slot in result['unavailable_slots']:
        print(f"  - {slot['start']} to {slot['end']}")
else:
    print("⛔ **Unavailable Slots:** None")
```
